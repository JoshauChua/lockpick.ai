```python
from bs4 import BeautifulSoup
import pandas as pd
import csv
import requests
import re

df = pd.DataFrame(data=None, columns=["Name", "type", "abilitie(s)", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed", "Weight (Kg)", "Height (m)"])
# for gen 1 only
dff = pd.DataFrame(data=None, columns=["Name", "type", "abilitie(s)", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed", "Weight (Kg)", "Height (m)"])    
# for all gens
```

```python

url = "https://pokeapi.co/api/v2"
def pokemon_info(pokeName):
    new_url = f"{url}/pokemon/{pokeName}"
    response = requests.get(new_url)
    if response.status_code == 200:
        try:
            return response.json()
        except ValueError:
            print(f"Error decoding JSON for {pokeName}")
            return None
    else:
        print(f"Failed to fetch data for {pokeName}: {response.status_code}")
        return None

gen1_pokemon = [
    "bulbasaur","ivysaur","venusaur",
    "charmander","charmeleon","charizard",
    "squirtle","wartortle","blastoise",
    "caterpie","metapod","butterfree",
    "weedle","kakuna","beedrill",
    "pidgey","pidgeotto","pidgeot",
    "rattata","raticate",
    "spearow","fearow",
    "ekans","arbok",
    "pikachu","raichu",
    "sandshrew","sandslash",
    "nidorina","nidoqueen","nidorino","nidoking",
    "clefairy","clefable",
    "vulpix","ninetales",
    "jigglypuff","wigglytuff",
    "zubat","golbat",
    "oddish","gloom","vileplume",
    "paras","parasect",
    "venonat","venomoth",
    "diglett","dugtrio",
    "meowth","persian",
    "psyduck","golduck",
    "mankey","primeape",
    "growlithe","arcanine",
    "poliwag","poliwhirl","poliwrath",
    "abra","kadabra","alakazam",
    "machop","machoke","machamp",
    "bellsprout","weepinbell","victreebel",
    "tentacool","tentacruel",
    "geodude","graveler","golem",
    "ponyta","rapidash",
    "slowpoke","slowbro",
    "magnemite","magneton",
    "farfetchd",
    "doduo","dodrio",
    "seel","dewgong",
    "grimer","muk",
    "shellder","cloyster",
    "gastly","haunter","gengar",
    "onix",
    "drowzee","hypno",
    "krabby","kingler",
    "voltorb","electrode",
    "exeggcute","exeggutor",
    "cubone","marowak",
    "hitmonlee","hitmonchan",
    "lickitung",
    "koffing","weezing",
    "rhyhorn","rhydon",
    "chansey",
    "tangela",
    "kangaskhan",
    "horsea","seadra",
    "goldeen","seaking",
    "staryu","starmie",
    "mr. mime",
    "scyther",
    "jynx",
    "electabuzz",
    "magmar",
    "pinsir",
    "tauros",
    "magikarp","gyarados",
    "lapras",
    "ditto",
    "eevee","vaporeon","jolteon","flareon",
    "porygon",
    "omanyte","omastar",
    "kabuto","kabutops"
    ]

for pokemon in gen1_pokemon:
    current_data = pokemon_info(pokemon)
    if current_data:  # Only process if data is valid
        try:
            necc_data = [
                current_data["name"],
                ", ".join([t["type"]["name"] for t in current_data["types"]]),
                ", ".join([a["ability"]["name"] for a in current_data["abilities"]]),
                current_data["stats"][0]["base_stat"],  # HP
                current_data["stats"][1]["base_stat"],  # Attack
                current_data["stats"][2]["base_stat"],  # Defense
                current_data["stats"][3]["base_stat"],  # Sp. Atk
                current_data["stats"][4]["base_stat"],  # Sp. Def
                current_data["stats"][5]["base_stat"],  # Speed
                current_data["weight"],  # Weight (Kg)
                current_data["height"] / 10.0,  # Height (m)
            ]
            df.loc[len(df)] = necc_data
        except KeyError as e:
            print(f"Missing data for {pokemon}: {e}")
```

```python
print(df)
df.to_csv("gen1_stats_(excluding_nidorans_and_mr.mime).csv", index=False)
```

```python


huge_url = "https://pokeapi.co/api/v2/pokemon?limit=20000"
big_data = requests.get(huge_url).json()

all_pokemon = [entry["name"] for entry in big_data["results"]] #every pokemon

allowed = re.compile(r'^[a-z-]+$') #list of allowed ASCII characters
allowed_pokemon = [name for name in all_pokemon if allowed.match(name)]


for pokemon in allowed_pokemon:
    current_data = pokemon_info(pokemon)
    if current_data:  # Only process if data is valid
        try:
            necc_data = [
                current_data["name"],
                ", ".join([t["type"]["name"] for t in current_data["types"]]),
                ", ".join([a["ability"]["name"] for a in current_data["abilities"]]),
                current_data["stats"][0]["base_stat"],  # HP
                current_data["stats"][1]["base_stat"],  # Attack
                current_data["stats"][2]["base_stat"],  # Defense
                current_data["stats"][3]["base_stat"],  # Sp. Atk
                current_data["stats"][4]["base_stat"],  # Sp. Def
                current_data["stats"][5]["base_stat"],  # Speed
                current_data["weight"],  # Weight (Kg)
                current_data["height"] / 10.0,  # Height (m)
            ]
            dff.loc[len(dff)] = necc_data
        except KeyError as e:
            print(f"Missing data for {pokemon}: {e}")

```
